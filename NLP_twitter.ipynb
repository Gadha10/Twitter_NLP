```python
df['Target']=df['Target'].map({'Positive':1,'Negative':-1,'Neutral':0})
```

```python
df
```

```python
tweets=df.Text
tweets
```

```python
#removing special characters
tweets=tweets.str.replace('[^a-zA-Z0-9]+'," ")
tweets
```

```python
stemmer=SnowballStemmer('english')
tk=TweetTokenizer()
#tokenize-lowercase-stemming
tweets=tweets.apply(lambda x:[stemmer.stem(i.lower()) for i in tk.tokenize(x)]).apply(lambda y:" ".join(y))
tweets
```

```python
#importing stopwords
from nltk.corpus import stopwords
nltk.download('stopwords')
sw=stopwords.words('english')
print(sw)
```

```python
#removing stop words
tweets=tweets.apply(lambda x:[i for i in tk.tokenize(x) if i not in sw]).apply(lambda y:" ".join(y))
tweets
```

```python
#vectorizing
vectorizer=TfidfVectorizer()
train_vec=vectorizer.fit_transform(tweets)
print(train_vec)
```

```python
X=train_vec
y=df['Target'].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
sv=SVC()
mn=MultinomialNB()
rf=RandomForestClassifier()
ab=AdaBoostClassifier()
lst=[sv,mn,rf,ab]
```

```python
for model in lst:
  print("\n model---",model)
  model.fit(X_train,y_train)
  y_pred=model.predict(X_test)
  print("Accuracy:",accuracy_score(y_test,y_pred)*100)
 #print("Classification report:",classification_report(y_test,y_pred))
 #print(ConfusionMatrixDisplay(y_test,y_pred)) 
  print("__"*100)
```

SVM has the highest accuracy compared to other three models.
